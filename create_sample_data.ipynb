```python
def create_random_date():
    start_date = datetime.date(2000, 1, 1)
    end_date = datetime.date.today()
    time_between_dates = end_date - start_date
    days_between_dates = time_between_dates.days
    random_number_of_days = random.randrange(days_between_dates)
    random_date = start_date + datetime.timedelta(days=random_number_of_days)
    return random_date

def create_uuid():
    return uuid.uuid4().hex[:12] 

def create_rating():
    return int(random.random()*5 + 1)
```

```python
# create db for restaurants

from tqdm import tqdm

def create_n_restaurants(n):
    restaurant_list = []
    for i in tqdm(range(n)):
        r = create_random_restaurant()
    #     print('adding restaurant: ', r.name)
        restaurant_list.append([r.id, r.name, r.zip_code, r.lat_long, r.date_added])
    restaurant_df_columns = ['restaurant_id', 'restaurant_name', 'restaurant_zipcode','restaurant_lat_long','restaurant_date_added']
    restaurants_df = pd.DataFrame(restaurant_list, columns=restaurant_df_columns)
    return restaurants_df

def create_n_dishes(restaurants_df,n):
    # create db for dishes at restaurants
    dishes_list = []
    for i in tqdm(range(n)):
        restaurant_id = restaurants_df['restaurant_id'].sample().values[0]
        d = create_random_dish(restaurant_id)
        dishes_list.append([d.id, d.name, d.restaurant_id, d.rating, d.num_reviews, d.date_added])
    dishes_df_columns = ['dish_id', 'dish_name', 'restaurant_id','dish_rating','dish_num_reviews','dish_date_added']
    dishes_df = pd.DataFrame(dishes_list, columns=dishes_df_columns)
    return dishes_df

def create_n_users(n):
    # create db for users
    users_list = []
    for i in tqdm(range(100)):
        u = create_random_user()
        users_list.append([u.id, u.first_name, u.last_name, u.user_rating, u.user_date_added, u.user_date_last_login, u.user_zipcode, u.user_lat_long])
    users_df_columns = ['user_id', 'first_name', 'last_name','user_rating','user_date_added','user_date_last_login','user_zipcode','user_lat_long']
    users_df = pd.DataFrame(users_list, columns=users_df_columns)
    return users_df
    
```
